</p>

<p style="font-family: Georgia; font-size: 25px; color: black;">
 Remove unusefull column
</p>

```python
df.drop('customer_unique_id',axis=1,inplace=True)
```

<p style="font-family: Georgia; font-size: 25px; color: black;">Convert Required colmns data typeÂ¶</p>

```python
# def convert_date_format(x):
#     try:
#         dt = pd.to_datetime(x, format="%d-%m-%y-%H-%M", errors='coerce')
#         return dt.strftime("%Y-%m-%d-%H-%M") if pd.notnull(dt) else np.nan
#     except:
#         return np.nan


date_cols = ['order_purchase_timestamp','order_approved_at','order_delivered_carrier_date',
                'order_delivered_customer_date','order_estimated_delivery_date',
                'shipping_limit_date','review_creation_date','review_answer_timestamp']


for col in date_cols:
    df[col] = df[col].astype(str).str.strip()
    df[col] = df[col].replace(['', 'nan', 'NaN', 'None'], np.nan)
    df[col] = pd.to_datetime(df[col], format="%d-%m-%Y %H:%M", errors='coerce',dayfirst=True)


        
others_cols=['review_score','payment_installments','payment_sequential','seller_zip_code_prefix','product_name_lenght','product_description_lenght','product_photos_qty','order_item_id']
df[others_cols]=df[others_cols].astype('Int64')
```

<p style="font-family: Georgia; font-size: 15px; color:green;">Converts multiple date-related columns from string format into datetime objects for accurate date/time calculations </p>

<p style="font-family: Georgia; font-size: 25px; color: black;">
Finding missing values
</p>

```python
df.isnull().sum()
```

<p style="font-family: Georgia; font-size: 25px; color: black;">
 Filling Missing Values
</p>

```python
df['order_approved_at']= df['order_approved_at'].fillna(df['order_purchase_timestamp'])
df['order_delivered_carrier_date']= df['order_delivered_carrier_date'].fillna(df['order_approved_at'])
df['order_delivered_customer_date']= df['order_delivered_customer_date'].fillna(df['order_estimated_delivery_date'])
df['shipping_limit_date']= df['shipping_limit_date'].fillna(df['order_purchase_timestamp'])
df['review_creation_date']= df['review_creation_date'].fillna(df['order_estimated_delivery_date'])
df['review_answer_timestamp']= df['review_answer_timestamp'].fillna(df['review_creation_date'])

###numeric columns filing with mean
int_cols   = ['product_name_lenght','product_description_lenght',
              'product_photos_qty','payment_sequential','payment_installments','order_item_id','seller_zip_code_prefix', 'review_score']
df[int_cols]   = df[int_cols].fillna(df[int_cols].median())

float_cols = ['price','freight_value','payment_value','product_weight_g','product_length_cm',
              'product_height_cm','product_width_cm']

df[float_cols] = df[float_cols].fillna(df[float_cols].mean())
# ##dropna which columns null values minimum
# df.dropna(subset=[',, ,, ,],inplace=True)

col=['product_category_name','product_category_name_english','seller_city','seller_state','payment_type','product_id','seller_id','review_id']

df[col]=df[col].ffill()

df['review_comment_message']=df['review_comment_message'].fillna('No Massage')

df['review_comment_title']=df['review_comment_title'].fillna('No Title')
```

<p style="font-family: Georgia; font-size: 30px; color: darkblue;">
ðŸ”¹Feature Engneering
</p>

<p style="font-family: Georgia; font-size: 25px; color: black;">
Create Month year colmun
</p>

```python
df['order_month']=df['order_purchase_timestamp'].dt.month
df['order_Year']=df['order_purchase_timestamp'].dt.year
df['order_month_name']=df['order_purchase_timestamp'].dt.month_name()
```

<p style="font-family: Georgia; font-size: 25px; color: black;">
Extract Delivery Duration
</p>

```python
df['delivery_time'] = (df['order_delivered_customer_date'] - df['order_purchase_timestamp']).dt.days
df['delivery_delay'] =( df['order_delivered_customer_date'] - df['order_estimated_delivery_date']).dt.days
```

<p style="font-family: Georgia; font-size: 25px; color: black;">
Profit Calculation
</p>

```python
# Total Cost = price + freight
df['total_cost'] = df['price'] + df['freight_value']

# Profit
df['profit'] = df['payment_value'] - df['total_cost']

# Profit Margin
df['profit_margin'] =np.where(df['payment_value']>0,(df['profit'] / df['payment_value']) * 100,np.nan)
```

Show the monthly profit trend

```python
# Define month order
month_order = ['January', 'February', 'March', 'April', 'May', 'June',
               'July', 'August', 'September', 'October', 'November', 'December']

# Convert to categorical with order
df['order_month_name'] = pd.Categorical(df['order_month_name'], categories=month_order, ordered=True)

# Then group and sort
monthly_profit = df.groupby('order_month_name',observed=True)['profit'].sum().sort_index()
plt.figure(figsize=(12,6))
plt.plot(monthly_profit.index,monthly_profit.values,marker='o',color='orange')
plt.xlabel('order_month')
plt.ylabel('Total_Profit')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```
